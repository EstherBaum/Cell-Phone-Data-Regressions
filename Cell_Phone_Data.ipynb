```python
# Statistical Analysiswork for Final Project

import pandas as pd
import numpy as np
import scipy.stats as stats
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import matplotlib.lines as mlines
```

```python
# Uploading a file to google colab
from google.colab import files

uploaded = files.upload()
```

```python
# import dataset
df = pd.read_csv("Cell_Phone_Use_Survey.csv")

df.head()
```

```python
# Scatterplot - Phone Pickups vs Notifications

X = df["Notifications"].values.reshape(-1, 1)
y = df["Phone Pickups"]

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)


# Predict the values using the model (this gives the regression line)
y_pred = model.predict(X)

# Calculate the Pearson correlation coefficient (r-value)
r_value, p_value = pearsonr(df["Notifications"], df["Phone Pickups"])

# Plot the data points
plt.scatter(X, y, color='blue', label='Notifications vs Phone Pickups')

# Plot the regression line
plt.plot(X, y_pred, color='red', label='Linear Regression Line')

# Add title and labels
plt.title('Notifications Received vs Phone Pickups')
plt.xlabel('Notifications Received')
plt.ylabel('Phone Pickups')

# Create custom legend entry for r-value and p-value
dummy_line = mlines.Line2D([], [], color='none', label=f"r-value: {round(r_value, 4)}, p-value: {round(p_value, 4)}")

# Add the custom legend with both the regression line and the r-value/p-value
plt.legend(handles=[dummy_line, plt.Line2D([0], [0], color='red', lw=2, label='Linear Regression Line')])

# Save PDF of Graph
plt.savefig('notifications_vs_phone_pickups.png', format='png')

# Show the plot
plt.show()

# Show r-value and p_value
print(f"r-value is: {round(r_value,4)}")
print(f"p-value is: {round(p_value,4)}")
```
